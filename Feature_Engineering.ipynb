6. **What are continuous and categorical variables? How do we handle categorical variables in Machine Learning? What are the common techniques?**

    * Continuous variables- These variable can take any value within a given range e.g., height, temperature.
    * Categorical variables-  Thes evariable represent characteristics or qualities that can be grouped into distinct categories e.g., colors, gender, ship mode etc.

      **Common techniques are:**
        * **Label Encoding**- Label Encoding assigns a unique integer to each category. It's best used for ordinal categorical variables, where there is a clear order or hierarchy between the categories.

        **Example**: For a variable like "T-shirt size" with categories S, M, L, and XL, you would encode them as:

                  S -> 1, M -> 2, L -> 3, XL -> 4

      * **One-Hot Encoding**- One-Hot Encoding creates a new binary column for each category. It is the most common technique and is ideal for nominal categorical variables, where there is no inherent order.

      **Example**: For a variable like "Color" with categories Red, Blue, Green:

          Red becomes [1, 0, 0]
          Blue becomes [0, 1, 0]
          Green becomes [0, 0, 1]

      * **Dummy Encoding**- Dummy Encoding is a variation of one-hot encoding. It's a bit more efficient as it creates k-1 new columns for k categories.

        **Example**: For the same "Color" variable with categories Red, Blue, Green, we can drop one category (e.g., Red) and represent it with zeros.

              Red becomes [0, 0]
              Blue becomes [1, 0]
              Green becomes [0, 1]

7. **What do you mean by training and testing a dataset?**
  > These are the steps used to develop and evaluate the machine learning models. This process involves dividing a complete dataset into two distinct subsets:
      * Training dataset- This subset is used to train the machine learning model. It comprises 70-80% of overall dataset.
      * Testing dataset- This dataset is used to evaluate the performance of the trained model on unseen data. After the model has learned from the training data, it is presented with the testing data, and its predictions are compared to the actual known outcomes in the testing set. It is smaller in portion approx 20-30%.

8. **What is sklearn.preprocessing?**
  > It is a module in scikit-learn which provides methods and transformer classes whuch are designed to prepare raw feature vectors which can be use with machine learning estimators. This process is know as data preprocessing.
  Some fearures are :
* **Scaling and Standardization**:
        * StandardScaler: Removes the mean and scales to unit variance, suitable for data without significant outliers.
        * MinMaxScaler: Scales features to a specified range, typically between 0 and 1.
        * RobustScaler: Scales data using the median and interquartile range, making it more robust to outliers.
        * Normalizer: Scales individual samples to have unit norm.

* **Encoding Categorical Features**:
        * OneHotEncoder: Transforms categorical features into a one-hot numerical representation.
        * LabelEncoder: Encodes target labels with values between 0 and 1.
        * TargetEncoder: Encodes categorical features based on the mean of the target variable for each category.
* **Feature Transformation and Generation**:
      * PolynomialFeatures: Generates polynomial features and interaction terms from existing features.
      * KBinsDiscretizer: Discretizes continuous features into fixed number of bins.
      * FunctionTransformer: Applies a custom function to transform data.
* **Imputation of Missing Values**:
      * SimpleImputer: Fills in missing values using strategies like mean, median, or most frequent.

10. **How do we split data for model fitting (training and testing) in Python? How do you approach a Machine Learning problem?**

```python
import seaborn as sns
from sklearn.model_selection import train_test_split

iris = sns.load_dataset("iris")
print("Iris dataset : \n")
print(iris.head())
print(iris.shape)
X = iris.drop("species", axis=1)
y = iris["species"]

print(f"\nDatset after removing target : \n{X.head()}")
print(f"\nTarget dataset : \n{y.head()}")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

14. **How can you find correlation between variables in Python?**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Variable_A': [10, 15, 20, 25, 30],
        'Variable_B': [2, 4, 5, 7, 8]}

df = pd.DataFrame(data)

correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

16. **What is an Optimizer? What are different types of optimizers? Explain each with an example.**
  > The optimizer is a crucial element in the learning process of the ML model. It is an algorithm that adjusts a machine learning model internal parameters (weights and biases) to minimize a loss function, which measures the error in the models predictions, thereby improving the model overall performance and accuracy.

  **Types of Optimizers:**
    * **Gradient Descent (GD)**- This is the most basic optimizer. It computes the gradient (the slope) of the loss function using the entire training dataset for each update step.

      **Example**-
      Let we have a dataset of 100,000 house prices. To update the model's weights, Gradient Descent would calculate the error for all 100,000 houses, average them, and then make one single update to the model.

  * **Stochastic Gradient Descent (SGD)**- SGD is an improvement on GD. Instead of using the entire dataset, it calculates the gradient and updates the parameters using just one randomly selected data point at a time.

      **Example**- For the same 100,000-house dataset, SGD would pick one house at random, calculate the error, update the weights, then pick another random house and repeat. This process is much faster.
    
  * **Adam (Adaptive Moment Estimation)**- Adam is one of the most popular and effective optimizers today. It's an adaptive learning rate optimizer that combines the strengths of two other optimizers: RMSprop and momentum.
  It quickly finds good weights for the millions of parameters, often without much manual hyperparameter tuning.

25. **Explain data encoding?**
  > Data encoding is the process of converting data from one format or structure to another. Simply means, it is used to handle categorical data, which represents labels or categories rather than numerical values. Most machine learning algorithms are designed to work with numerical input.

  > **Ways to do data encoding**:
    * **Label Encoding**: Assigns a unique integer to each category. For example, "red" might be 0, "green" 1, and "blue" 2. .
    * **One-Hot Encoding**: Transforms each category into a binary vector. For a categorical feature with 'n' unique categories, it creates 'n' new binary columns. For example, if "color" has "red," "green," and "blue," one-hot encoding creates three columns: is_red, is_green, and is_blue, with a '1' in the corresponding column and '0' in others.
    * **Ordinal Encoding**: Used for ordinal data, where categories have a natural order (e.g., "low," "medium," "high"). It assigns numerical values that reflect this order (e.g., "low" = 1, "medium" = 2, "high" = 3).
    * **Target Encoding**: Replaces each category with a statistical measure of the target variable for that category (e.g., the mean of the target variable for all instances belonging to that category).
