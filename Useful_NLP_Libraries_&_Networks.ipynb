**Q1) NLTK vs spaCy?**

>  NLTK and spaCy are both NLP libraries, but they differ in design and usage.

> NLTK is mainly used for learning and research. It provides many tools like tokenization, stemming, and tagging, but it is slower and requires more manual work.

> spaCy is designed for real-world applications. It is faster, more efficient, and provides built-in pipelines.

| Feature         | NLTK     | spaCy      |
| --------------- | -------- | ---------- |
| Ease of use     | Moderate | Easy       |
| Speed           | Slow     | Fast       |
| Use case        | Learning | Production |
| Built-in models | Limited  | Strong     |

**Q7) Tokenization + Frequency (NLTK)**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import re

nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field..."""

text = text.lower()
text = re.sub(r'[^\w\s]', '', text)

tokens = word_tokenize(text)
fdist = FreqDist(tokens)

print("Top words:")
print(fdist.most_common(5))
```

**Q8) LSTM Model using Keras**

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Data
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]

labels = [1, 1, 0, 0, 1]

# Convert labels
labels = np.array(labels)

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding
padded = pad_sequences(sequences, padding='post')

# Vocabulary size
vocab_size = len(tokenizer.word_index) + 1

# Model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=8))
model.add(LSTM(10))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train
model.fit(padded, labels, epochs=10, verbose=1)

print("Training successful ✅")
```

**Q9) spaCy NLP Pipeline**

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """Homi Jehangir Bhaba was an Indian nuclear physicist..."""

doc = nlp(text)

print("Tokens and Lemmas:")
for token in doc:
    print(token.text, "->", token.lemma_)

print("\nEntities:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

**Q10) Chatbot using LSTM/GRU + NLP**

 > To build a mental health chatbot:

- Step 1: Data Collection: Collect conversations and labeled intents.

- Step 2:
    - Preprocessing
    - Tokenization (spaCy)
    - Stopword removal
    - Lemmatization
- Step 3:
    - Model
    - Use LSTM/GRU to understand sequences
    - Input = sentence
    - Output = intent or response
- Step 4:
    - Response Generation
    - Use predefined responses or generative model

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Step 1: Data
texts = ["I feel sad", "I am happy", "I am stressed"]
labels = [0, 1, 0]   # 0 = negative, 1 = positive

labels = np.array(labels)

# Step 2: Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Step 3: Padding
max_len = max(len(seq) for seq in sequences)
padded = pad_sequences(sequences, maxlen=max_len, padding='post')

# Step 4: Vocabulary size
vocab_size = len(tokenizer.word_index) + 1

# Step 5: Model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=8))
model.add(LSTM(10))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 6: Train
model.fit(padded, labels, epochs=20, verbose=0)

print("Model trained ✅")

# Step 7: Prediction function
def predict_text(user_input):
    seq = tokenizer.texts_to_sequences([user_input])
    padded_seq = pad_sequences(seq, maxlen=max_len, padding='post')

    pred = model.predict(padded_seq, verbose=0)[0][0]

    if pred > 0.5:
        return "😊 Positive response: Glad you're feeling good!"
    else:
        return "😟 Negative response: I'm here for you. Stay strong."

# Step 8: User Input Loop
while True:
    user_input = input("\nEnter your message (or type 'exit'): ")

    if user_input.lower() == "exit":
        print("Chat ended.")
        break

    response = predict_text(user_input)
    print("Bot:", response)
```
