```python
'''
1. Write a Python function that takes a list of numbers
as input and returns the sum of all even numbers in the list
'''
sum_even=lambda lst:sum([i for i in lst if i%2==0])
print(sum_even([7,5,2,7,8,4,15]))
```

```python
'''
2. Create a Python function that accepts a string and
returns the reverse of that string.
'''
string_reverse=lambda string:string[::-1]
print(string_reverse("python"))
```

```python
'''
3. Implement a Python function that takes a list of integers and returns
a new list containing the squares of each number.
'''
squared_of_number=lambda lst:list(map(lambda x:x**2,lst))
print(squared_of_number([1,2,3,4]))
```

```python
'''
4. Write a Python function that checks if a given number is prime
or not from 1 to 200
'''
import math as m
def isPrime(n):
     prime_number={i: True for i in range(n + 1)}
     prime_number[0],prime_number[1]=False,False
     for i in range(2,int(m.sqrt(n))+1):
         if prime_number[i]==1:
             for j in range(i*i,n+1,i):
                 prime_number[j]=False
     return prime_number


print(isPrime(201))
```

```python
'''
5. Create an iterator class in Python that generates the Fibonacci
sequence up to a specified number of terms.
'''
n=int(input("enter the n terms for series "))

def fibo(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b
f=fibo(n)
for i in range(n):
  print(next(f))
```

```python
'''
 6. Write a generator function in Python that yields the powers of
 2 up to a given exponent.
 '''
def power_of_two(n):
  while n>=0:
    yield 2**n
    n=n-1
n=int(input("enter the number "))
s=power_of_two(n)
for i in range(n+1):
  print(next(s))
```

```python
'''
8.Use a lambda function in Python to sort a list of tuples based on the
second element of each tuple.
'''
data=[(5,2),(2,1),(9,6)]
data.sort(key=lambda x:x[1])
print(data)
```

```python
'''
9.Write a Python program that uses `map()` to convert a list of temperatures
from Celsius to Fahrenheit.
'''

temp_in_celsius=[36,38,32,30,29,100]
temp_in_farenhite=list(map(lambda temp:(temp * 9/5) + 32,temp_in_celsius))
print(temp_in_farenhite)
```

```python
'''
10.Create a Python program that uses `filter()` to remove all the vowels
from a given string
'''
string=input("enter the string")
vowels='aeiou'
x="".join((filter(lambda x: x not in vowels,string)))
print(x)
```
