1. **What is hypothesis testing in statistics?**
  
 > It is an assumption or supposition about the population on the basis of sample information about the population parameters involved and to arrive at decision we make assumption is called a hypothesis.
  Key Concepts in Hypothesis Testing:
    * Null Hypothesis (H0): A statement that there is no significant difference or effect. It's the assumption we try to disprove.
    * Alternative Hypothesis (H1): A statement that contradicts the null hypothesis, suggesting a significant difference or effect.
    * Significance Level (∝): The probability of rejecting the null hypothesis when it is actually true (Type I error). Commonly set at 0.05 (5%).
    * Test Statistic: A value calculated from the sample data that is used to determine whether to reject the null hypothesis.
    * P-value: If the p-value is less than the significance level, the null hypothesis is rejected.



2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**

  > Null hypothesis (H0)- It is definite statement about the population parameter which is tested for possible rejection under the assumption that it is true. Usually, a null hypothesis is written as no difference.

        H0:μ=μ0

  On other hand alternate hypothesis (H1), is complementary to the null hypothesis. Usually, an alternative hypothesis is written as some difference.

  For example,

      if we want to test the null hypothesis that the population has a specified mean μ0, i.e. H0:μ =μ0 , then the alternative hypothesis could be

      * H0:μ ≠ μ0   (2-tailed test)
      * H0:μ > μ0  (right one tailed)
      * H0:μ <μ0  (left one tailed)


3. **Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**
  
  > Significance level ∝ tell us how much sure we want to be before saying the claim is false. Actually it is the confidence with which we reject or accept the null hypothesis. Usually, we choose 0.05 (5%).

5. **What is the difference between a Z-test and a T-test? Explain when to use each.**

  > Z-tests and t-tests are both statistical methods used to compare means, but they differ in their assumptions.
    * Z-tests are used when the population standard deviation is known, or when the sample size is large (generally > 30).
    * T-tests are used when the population standard deviation is unknown and the sample size is small (generally < 30).

```python
"""
7.  Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1,
49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
"""

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1,
               49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume mean of population
mu_0 = 50

mean_sample = np.mean(sample_data)
std_sample = np.std(sample_data, ddof=0)  # population std
n = len(sample_data)

z_stat = (mean_sample - mu_0) / (std_sample / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean = {mean_sample:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
"""
8. Write a Python script to simulate data from a normal distribution and calculate
the 95% confidence interval for its mean. Plot the data using Matplotlib.
"""

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(0)

mean=20
std_dev=5
sample=100


data=np.random.normal(mean,std_dev,sample)

n=len(data)

mean=np.mean(data)
sample_std=np.std(data,ddof=1)
standard_error=sample_std/np.sqrt(sample)

confidence=0.95
degree_of_freedom=sample-1

confidence_interval=stats.t.interval(confidence,degree_of_freedom,mean,standard_error)


print(f"mean : {mean:.2f}")
print(f"sample_std : {sample_std:.2f}")
print(f"standard_error : {standard_error:.2f}")
print(f"confidence_interval : {confidence_interval}")

plt.hist(data,edgecolor="k",align="left",facecolor="lightblue")
plt.title("Histogram Normal Distribution ")
plt.axvline(mean,color="r")
plt.axvline(confidence_interval[0],color="g")
plt.axvline(confidence_interval[1],color="g")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.show()
```

```python
"""
9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores
represent in terms of standard deviations from the mean.
"""
import numpy as np
import matplotlib.pyplot as plt

def zGraph(data):

    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)

    z_scores = (data - mean) / std_dev

    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1)
    plt.show()

    return z_scores

sample_data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9])

z_scores = zGraph(sample_data)
print(z_scores)
```
