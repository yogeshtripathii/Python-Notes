```python
'''
Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and
standard deviation using NumPy, and draw a histogram to visualize the
distribution.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean = 50
std_dev = 5
data = norm.rvs(loc=mean, scale=std_dev, size=1000)
computed_mean = np.mean(data)
computed_std_dev = np.std(data)
print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std_dev:.2f}")

plt.hist(data, bins=10, edgecolor='black')
plt.show()
```

```python
'''
Question 10:
You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall
sales trend. '
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
â—	Explain how you would apply the Central Limit Theorem to estimate the average
sales with a 95% confidence interval.
â—	Write the Python code to compute the mean sales and its confidence interval.
'''

'''
The Central Limit Theorem (CLT) states that if we repeatedly take sufficiently
large random samples from a population, the distribution of the sample means
will approximate a normal distribution, regardless of the population's original
distribution. This also holds approximately for smaller samples if the
underlying data is not heavily skewed.

In this problem, we have a sample of daily sales data. Although our sample size
(n = 20) is less than 30, we can still use the CLT concept but must use the
t-distribution instead of the normal distribution to account for additional
uncertainty in estimating the standard deviation.
'''

import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.array(daily_sales)

mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)
n = len(data)

standard_error = std_sales / np.sqrt(n)

t_crit = stats.t.ppf(1 - 0.025, df=n-1)

ci_lower = mean_sales - t_crit * standard_error
ci_upper = mean_sales + t_crit * standard_error

print(__doc__)
print(f"mean_sales : {mean_sales}, C.I : {ci_lower, ci_upper}")

print('''
We are 95% confident that the true average daily sales for the company lies
between 240.169 and $256.33
''')
```
